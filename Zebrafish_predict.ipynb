Select the minimum and maximum of all images 

```python
minimum=1000
maximum=0
selected_slice=40
j=0
for i in range(selected_slice,len(dir),100):
    im=h5py.File('data/'+dir[i],'r+')
    img=list(im['d1']) 
    new_min=np.min(img)
    new_max=np.max(img)
    if new_min < minimum:
        minimum=new_min
        print (minimum)
    if new_max > maximum:
        maximum=new_max
        print (maximum)
    j=j+1
    print (dir[i])
    
num_img=j
    
    
```

Convert the image values to  be between -1 and 1

```python
data=np.zeros((num_img,400,400))
j=0

# Data starts by having values between minimum and maximum
for i in range (selected_slice,len(dir),100):
    im=h5py.File('data/'+dir[i],'r+')
    img=list(im['d1']) 
    img=img-minimum # So the data is between 0 and maximum
    img=img/(0.5*(maximum-minimum)) # So the data is between 0 and 2
    img=img-1 # So the data is between -1 and 1
    data[j,:,:]=img # Store the image into the variable data
    j=j+1
    
```

Create the downsampled version of the images

```python
down_data=np.zeros((num_img,8,8))
for i in range(0,len(down_data)):
    img=data[i,:,:]
    dimg=np.zeros((8,8))
    for j in range (0,8):
        for k in range (0,8):
            dimg[j,k]=np.mean(img[j*50:j*50+50,k*50:k*50+50])
    down_data[i,:,:]=dimg
```

Create the labels for each image

```python
labels=np.zeros((num_img))
count0=0
count1=0
for i in range (0,len(down_data)):
    img=down_data[i,:,:]
    left_max=np.max(img[4:8,:])
    right_max=np.max(img[0:4,:])
    if left_max > right_max:
        labels[i]=0
        count0=count0+1
    else:
        labels[i]=1
        count1=count1+1
```

## Model

```python
import tensorflow as tf
import keras
from keras import layers
from keras import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers import Dropout
```

```python
num_samples=len(y_train)
num_timestep=3
num_features=25
x_train = x_train.reshape(x_train.shape[0], 8, 8, 1)
x_test = x_test.reshape(x_test.shape[0], 8, 8, 1)
```

```python
model = Sequential()
model.add(Conv2D(64, kernel_size=(3, 3),activation='relu',input_shape=(8,8,1),data_format='channels_last'))  
model.add(Conv2D(64, (3, 3), activation='relu'))
#model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['acc'])
```

```python
history=model.fit(x_train,y_train,epochs=50)
```
