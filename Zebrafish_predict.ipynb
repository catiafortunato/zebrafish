## Downsample the images

Select the directory

```python
dir=os.listdir('/home/catia/Desktop/zebrafish/data')
```

```python
dir.sort()
```

Select the minimum and maximum of all images 

```python
minimum=1000
maximum=0
selected_slice=40
j=0
for i in range(selected_slice,len(dir),100):
    im=h5py.File('data/'+dir[i],'r+')
    img=list(im['d1']) 
    print (dir[i])
    new_min=np.min(img)
    new_max=np.max(img)
    if new_min < minimum:
        minimum=new_min
        print (minimum)
    if new_max > maximum:
        maximum=new_max
        print (maximum)
    j=j+1
    
    
num_img=j
```

Convert the image values to  be between -1 and 1

```python
data=np.zeros((num_img,400,400))
j=0

# Data starts by having values between minimum and maximum
for i in range (selected_slice,len(dir),100):
    im=h5py.File('data/'+dir[i],'r+')
    img=list(im['d1']) 
    img=img-minimum # So the data is between 0 and maximum
    img=img/(0.5*(maximum-minimum)) # So the data is between 0 and 2
    img=img-1 # So the data is between -1 and 1
    data[j,:,:]=img # Store the image into the variable data
    j=j+1
    
```

**Test if the data is similar**, image on the right is the one obtained by drectly ploting the image in the directory, and the image on the left is the one obtained by ploting the version of the image with values between -1 and 1

```python
im=h5py.File('data/'+dir[40],'r+')
img=list(im['d1'])
mean=np.mean(img)

f = plt.figure()
f.add_subplot(1,2, 1)
plt.imshow(img)
f.add_subplot(1,2, 2)
plt.imshow(data[1,:,:])
plt.show(block=True)
```

Create the downsampled version of the images

```python
down_data=np.zeros((num_img,8,8))
for i in range(0,len(down_data)):
    img=data[i,:,:]
    dimg=np.zeros((8,8))
    for j in range (0,8):
        for k in range (0,8):
            dimg[j,k]=np.mean(img[j*50:j*50+50,k*50:k*50+50])
    down_data[i,:,:]=dimg
```

Create the labels for each image

```python
labels=np.zeros((num_img))
count0=0
count1=0
for i in range (0,len(down_data)):
    img=down_data[i,:,:]
    left_max=np.max(img[4:8,0:5])
    right_max=np.max(img[0:4,0:5])
    if left_max > right_max:
        labels[i]=0
        count0=count0+1
    else:
        labels[i]=1
        count1=count1+1
```

## Model

```python
import tensorflow as tf
import keras
from keras import layers
from keras import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers import Dropout
from keras.layers import BatchNormalization
from keras import regularizers
```

```python
num_samples=len(y_train)
num_timestep=3
num_features=25
x_train = x_train.reshape(x_train.shape[0], 8, 8, 1)
x_test = x_test.reshape(x_test.shape[0], 8, 8, 1)
```

```python
model = Sequential()
model.add(Conv2D(16, kernel_size=(3, 3),activation='relu',input_shape=(8,8,1),data_format='channels_last',))  
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(lr=0.00001),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['acc'])
```

```python
history=model.fit(x_train,y_train,epochs=100,validation_split=0.2)
```

```python
history_dict=history.history
plt.clf()
acc_values = history_dict['acc']
val_acc_values = history_dict['val_acc']

epoch=range(1, 100 + 1)

plt.plot(epoch, acc_values, 'bo', label='Training acc')
plt.plot(epoch, val_acc_values, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

From this plot we can see that the training accuracy increaces with time, in the first steps it is constant at value 0.8696 due to the unbalanced data, but reaches an accuracy of 0.93.

Besides that the validation accuracy increases over time but is osccilating. This can mean that a portion of the validation examples are being classified randonmly. This can be a form of overfitting, where the model learns noise instead of the underlying relation.
